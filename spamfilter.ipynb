```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import io
```

```python
from google.colab import files
uploaded = files.upload()  # Remove the file path argument

# Get the first filename from the uploaded dictionary
filename = list(uploaded.keys())[0]

# Now read the CSV using the filename
# Use the filename variable as the key to access the uploaded dictionary
df = pd.read_csv(io.BytesIO(uploaded[filename]))
```

```python
print(df)
```

```python
data=df.where((pd.notnull(df)),'')
```

```python
data.head(10)
data.info
```

```python
data.shape
data.loc[data['Category'] == 'spam','Category']==0
data.loc[data['Category'] == 'ham','Category']==1
```

```python
X=data['Message']
Y=data['Category']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

```python
print(Y.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
feature_extraction=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train_features=feature_extraction.fit_transform(X_train)
X_test_features=feature_extraction.transform(X_test)
# Replace 'spam' and 'ham' with 0 and 1 in Y_train and Y_test
Y_train = Y_train.replace({'spam': 0, 'ham': 1}).astype(int)
Y_test = Y_test.replace({'spam': 0, 'ham': 1}).astype(int)
```

```python
print(X_train)
print(X_train_features)
```

```python
model=LogisticRegression()
model.fit(X_train_features,Y_train)
```

```python
prediction_on_training_data=model.predict(X_train_features)
accuracy_on_training_data=accuracy_score(Y_train,prediction_on_training_data)
```

```python
print('Accuracy on training data:',accuracy_on_training_data)
```

```python
prediction_on_test_data=model.predict(X_test_features)
accuracy_on_test_data=accuracy_score(Y_test,prediction_on_test_data)
```

```python
print('Accuracy on test data:',accuracy_on_test_data)
```

```python
input=['Hi. you have been selected for a loterry win for 10000000 dollars. click to access']
input_data_features=feature_extraction.transform(input)
prediction=model.predict(input_data_features)
print(prediction)
if(prediction[0]==1):
  print('Spam mail')
else:
  print('ham mail')
```
